# Compute Drawdowns

```python
import pandas as pd
import numpy as np
```

```python
me_m = pd.read_csv(
    "data/Portfolios_Formed_on_ME_monthly_EW.csv", 
    header=0,
    index_col=0,
    parse_dates=True,
    date_format="%Y%m",
    na_values=-99.99
)
rets = me_m[['Lo 10', 'Hi 10']]
rets.columns = ['SmallCap', 'LargeCap']
rets = rets / 100
rets.index = rets.index.to_period('M')
rets.head()
```

# Compute Drawdown

1. Compute a wealth index
2. Compute previous peaks
3. Compute drawdown which is the wealth as a percentage of the previous peaks

```python
wealth_index = 1000 * (1 + rets["LargeCap"]).cumprod()
wealth_index.head()
```

```python
wealth_index.plot.line()
```

```python
# The max seen so far
previous_peaks = wealth_index.cummax()
print(previous_peaks.head())
previous_peaks.plot.line()
```

```python
drawdown = (wealth_index - previous_peaks)/previous_peaks
print(drawdown.head())
drawdown.plot()
```

```python
drawdown.min()
```

```python
# Highest drwadown seen since 1975
drawdown.loc["1975":].min().item()
```

```python
drawdown.loc["1975":].idxmin(), drawdown.loc["1975":].min().item()
```

```python
drawdown.idxmin()
```

```python
def drawdown(return_series: pd.Series):
    """Takes a time series of asset returns.
       returns a DataFrame with columns for
       the wealth index, 
       the previous peaks, and 
       the percentage drawdown
    """
    wealth_index = 1000*(1+return_series).cumprod()
    previous_peaks = wealth_index.cummax()
    drawdowns = (wealth_index - previous_peaks)/previous_peaks
    return pd.DataFrame({"Wealth": wealth_index, 
                         "Peaks": previous_peaks, 
                         "Drawdown": drawdowns})

drawdown(rets["SmallCap"]).head()
```

```python
drawdowns_since_1975 = drawdown(rets["SmallCap"])["Drawdown"].loc["1975":]
drawdowns_since_1975.idxmin(), drawdowns_since_1975.min().item()
```
